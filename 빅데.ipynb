```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)

file_path = '/content/drive/MyDrive/데이터.xlsx'
data_sheet = pd.read_excel(file_path)

data_clean = data_sheet.copy()
data_clean.columns = data_clean.iloc[0]
data_clean = data_clean[1:].reset_index(drop=True)

new_columns = ['국적별(1)', '국적별(2)',
               '2021.10_계', '2021.10_관광', '2021.10_상용', '2021.10_공용', '2021.10_유학연수', '2021.10_기타',
               '2021.11_계', '2021.11_관광', '2021.11_상용', '2021.11_공용', '2021.11_유학연수', '2021.11_기타',
               '2021.12_계', '2021.12_관광', '2021.12_상용', '2021.12_공용', '2021.12_유학연수', '2021.12_기타']
data_clean.columns = new_columns

data_clean.iloc[:, 2:] = data_clean.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

monthly_sum = data_clean[['2021.10_계', '2021.11_계', '2021.12_계']].sum()

plt.figure(figsize=(8, 5))
bars = monthly_sum.plot(kind='bar', color=['skyblue', 'lightgreen', 'lightcoral'])

plt.title('월별 방문자 수 총합', fontproperties=font_prop, fontsize=14)
plt.xlabel('월별 항목', fontproperties=font_prop, fontsize=12)
plt.ylabel('방문자 수', fontproperties=font_prop, fontsize=12)
plt.xticks(ticks=range(len(monthly_sum.index)), labels=monthly_sum.index,
           fontproperties=font_prop, rotation=0, fontsize=10)
plt.yticks(fontproperties=font_prop, fontsize=10)

for bar, value in zip(bars.patches, monthly_sum):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 5000,
             f'{int(value):,}', ha='center', va='bottom', fontproperties=font_prop, fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)

data = {
    '월': ['10월', '10월', '10월', '11월', '11월', '11월', '12월', '12월', '12월'],
    '국적별(2)': ['한국', '중국', '일본', '한국', '중국', '일본', '한국', '중국', '일본'],
    '방문자수': [150, 100, 80, 200, 150, 90, 250, 180, 120]
}
purpose_summary = pd.DataFrame(data)

purpose_summary['방문자수'] = pd.to_numeric(purpose_summary['방문자수'], errors='coerce')

months = purpose_summary['월'].unique()
fig, axes = plt.subplots(1, len(months), figsize=(18, 6))
for i, month in enumerate(months):
    monthly_data = purpose_summary[purpose_summary['월'] == month]
    top_3 = monthly_data.nlargest(3, '방문자수')
    wedges, texts, autotexts = axes[i].pie(
        top_3['방문자수'],
        labels=top_3['국적별(2)'],
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors[:len(top_3)]
    )
    for text in texts + autotexts:
        text.set_fontproperties(font_prop)
    axes[i].set_title(f"{month} 방문자 분포", fontsize=14, fontproperties=font_prop)

plt.suptitle("국적별 월별 방문자 수 (상위 3개)", fontsize=16, fontproperties=font_prop)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)

data = {
    '국적': ['한국', '한국', '한국', '중국', '중국', '중국', '일본', '일본', '일본'],
    '방문목적': ['관광', '비즈니스', '유학', '관광', '비즈니스', '유학', '관광', '비즈니스', '유학'],
    '방문자수': [300, 150, 100, 200, 180, 80, 250, 140, 90]
}
df = pd.DataFrame(data)

pivot_table = df.pivot(index='방문목적', columns='국적', values='방문자수')

ax = pivot_table.plot(kind='bar', figsize=(14, 8), width=0.8, color=['skyblue', 'lightgreen', 'lightcoral'])

plt.title("국적별 방문 목적에 따른 방문자수 (상위 3개)", fontsize=16, fontproperties=font_prop)
plt.ylabel("방문자수", fontsize=12, fontproperties=font_prop)
plt.xlabel("방문 목적", fontsize=12, fontproperties=font_prop)

plt.xticks(rotation=0, fontsize=10, fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom',
                    fontsize=10, fontproperties=font_prop)

plt.legend(title="국적", prop=font_prop, title_fontproperties=font_prop)

plt.tight_layout()

plt.show()

```
